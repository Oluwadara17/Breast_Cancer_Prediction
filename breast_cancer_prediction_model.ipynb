```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

from sklearn.svm import SVC


from sklearn.feature_selection import RFE
```

```python
# Split the data into features (X) and target (y)
X = data_copy.drop(' Class', axis=1)
y = data_copy[' Class']

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1, stratify = y)
```

```python
# Initialize SVM and Random Forest algorithms
svm = SVC(kernel='linear')
rf = RandomForestClassifier(n_estimators=50, random_state=0)

# Perform RFE feature selection with SVM
svm_rfe = RFE(estimator=svm, n_features_to_select=9, step=1)
svm_rfe.fit(X_train, y_train)

# Perform RFE feature selection with Random Forest
rf_rfe = RFE(estimator=rf, n_features_to_select=9, step=1)
rf_rfe.fit(X_train, y_train)

# Print the selected feature rankings
print('Feature Rankings:', svm_rfe.ranking_)
print('Feature Rankings:', rf_rfe.ranking_)
print(" ")

# Print the selected features
selected_features = X_train.columns[svm_rfe.support_]
print('Selected Features for SVM:', selected_features)
print(" ")
selected_features = X_train.columns[rf_rfe.support_]
print('Selected Features for Random Forest:', selected_features)
```

```python
# Transform training and test sets to include only selected features
X_train_svm = svm_rfe.transform(X_train)
X_test_svm = svm_rfe.transform(X_test)

X_train_rf = rf_rfe.transform(X_train)
X_test_rf = rf_rfe.transform(X_test)
```

```python
# Train SVM and Random Forest using selected features
svm.fit(X_train_svm, y_train)
rf.fit(X_train_rf, y_train)
```

# Model Evaluation

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
# Predict on test set

svm_pred = svm.predict(X_test_svm)

print("Support Vector Machine(SVM) ")
print("accuracy is ",accuracy_score(svm_pred,y_test))
print ("precision is ",precision_score(svm_pred,y_test,pos_label=4))
print("recall is ",recall_score(svm_pred,y_test,pos_label=4))
print("f1 is ",f1_score(svm_pred,y_test,pos_label=4))
print(sns.heatmap(confusion_matrix(svm_pred, y_test),cmap='Accent', annot=True))
```

```python
# random forest 
rf_pred = rf.predict(X_test_rf)
print("Random Forest ")
print("accuracy is ",accuracy_score(rf_pred,y_test))
print ("precision is ",precision_score(rf_pred,y_test,pos_label=4))
print("recall is ",recall_score(rf_pred,y_test,pos_label=4))
print("f1 is ",f1_score(rf_pred,y_test,pos_label=4))
print(sns.heatmap(confusion_matrix(rf_pred, y_test),cmap='Accent', annot=True))
```
